## **Regressão Linear**
Considerando as fórmulas para os mínimos quadrados para os coeficientes da reta:




 




## Exemplo 1

Ajuste uma reta aos valores de x e y nas primeiras duas colunas da Tabela 17.1. E calcule o desvio padrão total, o erro padrão da estimativa e o coeficiente de correlação 



```python
# Algoritmo para ajuste de curvas por regressão linear
import numpy as np            # Importa módulo numpy para criação das matrizes e vetores

# Dados
x = [1, 2, 3, 4, 5, 6, 7]
y = [0.5, 2.5, 2.0, 4.0, 3.5, 6.0, 5.5]
n = 7


# Cálculo dos somatórios necessários
Soma_xy = 0
Soma_x = 0
Soma_y = 0
Soma_x2 = 0
Soma_y_y = 0
Erro = 0
for i in range(0,n):
    Soma_xy = Soma_xy + x[i]*y[i]
    Soma_x = Soma_x + x[i]
    Soma_y = Soma_y + y[i]
    Soma_x2 = Soma_x2 + x[i]*x[i]
       
y_med = Soma_y/n # Média de y
x_med = Soma_x/n # Média de x
a1 =  (n*Soma_xy - Soma_x*Soma_y) / (n*Soma_x2 - Soma_x**2) 
a0 = y_med - a1*x_med

for i in range(0,n):
    Soma_y_y = Soma_y_y + (y[i]-y_med)**2
    Erro = Erro + (y[i] - a0 -a1*x[i])**2

St = Soma_y_y
Sr = Erro
R2 = (St-Sr)/St

print('a0 = ', a0, ' a1 = ', a1)
print('R2 = ', R2)
  


```

## Exemplo 2

Ajuste a equação abaixo aos dados da Tabela 17.3 usando uma transformação logarítmica dos dados.



Considerando a equação a ser ajustada e sua linearização a seguir:



```python
# Algoritmo para ajuste de curvas por regressão linear
import numpy as np            # Importa módulo numpy para criação das matrizes e vetores

# Dados
dados_x = [1, 2, 3, 4, 5]
x = np.log10(dados_x)
print(x)
dados_y = [0.5, 1.7, 3.4, 5.7, 8.4]
y = np.log10(dados_y)
n = 5


# Cálculo dos somatórios necessários
Soma_xy = 0
Soma_x = 0
Soma_y = 0
Soma_x2 = 0
Soma_y_y = 0
Erro = 0
for i in range(0,n):
    Soma_xy = Soma_xy + x[i]*y[i]
    Soma_x = Soma_x + x[i]
    Soma_y = Soma_y + y[i]
    Soma_x2 = Soma_x2 + x[i]*x[i]
       
y_med = Soma_y/n # Média de y
x_med = Soma_x/n # Média de x
a1 =  (n*Soma_xy - Soma_x*Soma_y) / (n*Soma_x2 - Soma_x**2) 
a0 = y_med - a1*x_med

for i in range(0,n):
    Soma_y_y = Soma_y_y + (y[i]-y_med)**2
    Erro = Erro + (y[i] - a0 -a1*x[i])**2

St = Soma_y_y
Sr = Erro
R2 = (St-Sr)/St

print('a0 = ', a0, ' a1 = ', a1)
print('R2 = ', R2)
```
