## **Polinômio Interpolador de Lagrange**
O polinômio interpolador de Lagrange é uma reformulação do polinômio de Newton para evitar o cálculo de diferenças divididas.

## Exemplo 4

Use um polinômio interpolador de Lagrange de primeiro e de
segundo graus para calcular ln(2) com base nos dados fornecidos no Exemplo 1.




```python
# Algoritmo de interpolação de Lagrange
import numpy as np            # Importa módulo numpy para criação das matrizes e vetores

# Dados
x = np.array([1, 4, 6])
fx = np.array([0, 1.386294, 1.791760])

# Grau do polinômio
n = 2

# Ponto de interpolação
xx = 2

#Cálculo dos Coeficientes de Lagrange
L = np.ones(n+1)

f1 = 0
for i in range(n+1):
    prod = 1
    for j in range(n+1):
        if i!=j:
           prod = prod*(xx-x[j])/(x[i]-x[j])
    L[i] = prod
    f1 = f1 + L[i]*fx[i]
print(f1)
```

```python
# Algoritmo de interpolação de Lagrange - Scipy
import numpy as np                         # Importa módulo numpy para criação das matrizes e vetores
from scipy.interpolate import lagrange     # Importa lagrange de scipy para solução sistemas de equações lineares
import matplotlib.pyplot as plt

# Dados
x = np.array([1, 4, 6])
fx = np.array([0, 1.386294, 1.791760])

poly = lagrange(x, fx)
print(poly)
xp = np.arange(0,6,0.1)
yp = poly(xp)

plt.figure()
plt.plot(xp, yp)

plt.plot(x, fx, 'ro')
plt.show()

X = 2                                    # Valor a ser interpolado
print('f1(2) = ', poly(X))

erro_abs = np.log(2) - poly(X)
print('Erro absoluto = ', erro_abs)





```

## Exemplo:

Pode-se desenvolver um algoritmo para estudar o problema de análise de tendência associado com o paraquedista em queda livre. Suponha que tenhamos

Nosso problema é fazer uma estimativa da velocidade do paraquedista em t = 10 s para preencher o grande salto nas medidas entre t = 7 e t = 13 s. Estamos cientes de que o comportamento dos polinômios interpoladores pode ser inesperado. Logo, construiremos polinômios de quarto, terceiro, segundo e primeiro graus e compararemos os resultados.



```python
# Exemplo Slide Algoritmo de interpolação de Lagrange - Scipy
import numpy as np                         # Importa módulo numpy para criação das matrizes e vetores
from scipy.interpolate import lagrange     # Importa lagrange de scipy para solução sistemas de equações lineares
import matplotlib.pyplot as plt

# Dados
x = np.array([1, 3, 5, 7, 13])
fx = np.array([800, 2310, 3090, 3940, 4755])

poly = lagrange(x, fx)
print(poly)
xp = np.arange(0,14,0.1)
yp = poly(xp)

plt.figure()
plt.plot(xp, yp)

plt.plot(x, fx, 'ro')
plt.show()

X = 10
print('f1(10) = ', poly(X))
```

```python
# Exemplo Slide Algoritmo de interpolação de Lagrange - Scipy
import numpy as np                         # Importa módulo numpy para criação das matrizes e vetores
from scipy.interpolate import lagrange     # Importa lagrange de scipy para solução sistemas de equações lineares
import matplotlib.pyplot as plt

# Dados

x = np.array([9, 11 ])
fx = np.log10(x)

poly = lagrange(x, fx)
print(poly)
xp = np.arange(8,12,0.1)
yp = poly(xp)

plt.figure()
plt.plot(xp, yp)

plt.plot(x, fx, 'ro')
plt.show()

X = 10
print('Estimativa f1(10) = ', poly(X))

# Erro relativo percentual
erro = np.log10(10)-poly(X)
erro_relativo = erro/np.log10(10)
erro_relativo_percentual = erro_relativo*100
print('Erro relativo percentual =', erro_relativo_percentual)
```
